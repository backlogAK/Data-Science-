---------------------------- Defining axes, ticks, and grids --------------------------------

```python
# Basic Object-Oriented plotting
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.plot(x,y)
```

```python
# Adding ticks 

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])
ax.plot(x,y)
```

```python
# Add grid to background

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

---------------------------------- Generating multiple plots in one figure with subplots -------------------------

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2) # Using tuples (ax1, ax2)

ax1.plot(x)
ax2.plot(x,y)
```

------------------------- LABELS AND ANNOTATIONS ---------------------------

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

-------------------------- Labeling plot features ------------------------------

```python
# Functional method

x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]
plt.bar(x,y)
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
# same using pie-chart

z = [1,2,3,4,.5]
veh_type = ['bicycle','motor','car','van','stroller']

plt.pie(z,labels=veh_type)
plt.show()
```

```python
# Object-oriented method

cars = pd.read_csv('E:\mtcars.csv')
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')
```

--------------------------------- Add a Legend to your plot -------------------------------

```python
# The functional method

plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

```python
# The Object-Oriented method

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')
```

----------------------------- CREATING STATISTICAL DATA GRAPHICS -------------------------------------

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

```python
import seaborn as sb
sb.set_style('whitegrid')
```

---------------------------- Eyeballing dataset distributions with histograms ---------------------------

```python
cars = pd.read_csv('E:\mtcars.csv')
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

cars.index = cars.car_names

mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
# another way of creating histogram

plt.hist(mpg)
plt.plot()
```

```python
# Using seaborn to create histogram

sb.distplot(mpg)
```

------------------------- Seeing scatterplot in action --------------------------

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

```python
# making scatter plot using seaborn

sb.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

------------------------------ Generating a scatter plot matrix ------------------------------------

```python
sb.pairplot(cars)
```

```python
cars_subset = cars[['mpg','disp','hp','wt']]
sb.pairplot(cars_subset)
plt.show()
```

------------------------------ Building Boxplot -----------------------------------

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
# Using Seaborn

sb.boxplot(x='am', y='mpg', data=cars, palette='hls')
```
