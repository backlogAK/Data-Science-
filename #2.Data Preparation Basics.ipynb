Data Preparation Basics

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

-------------- Figuring Out What Data is Missing ------------

```python
missing = np.nan
series_obj = Series(['row 1','row 2',missing,'row 4','row 5','row 6',missing,'row 8'])
series_obj
```

```python
series_obj.isnull()
```

-------------- Filling in for missing values ------------

```python
# seed to get same number as you type
np.random.seed(25)
DF_obj = DataFrame(np.random.rand(36).reshape(6,6))
DF_obj
```

```python
# .loc method to select rows and columns and then give values to missing

DF_obj.loc[3:5,0] = missing
DF_obj.loc[1:4,5] = missing
DF_obj
```

```python
# filling value for NaN

filled_DF = DF_obj.fillna(0)
filled_DF
```

```python
filled_DF = DF_obj.fillna({0:0.1, 5:1.25})
filled_DF
```

```python
# filled with last non null element in the column series (Eg: in 2:0 is copied to 0 column missing, same for 0:5) 

filled_DF = DF_obj.fillna(method='ffill')
filled_DF
```

-------------- Counting missing values ------------

```python
np.random.seed(25)
DF_obj = DataFrame(np.random.rand(36).reshape(6,6))
DF_obj.loc[3:5,0] = missing
DF_obj.loc[1:4,5] = missing
DF_obj
```

```python
# for counting (Eg: column 0 has 3 null values, cloumn 5 has 4 null values)
DF_obj.isnull().sum()
```
